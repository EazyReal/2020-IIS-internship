```python
def rank_of_models(js, dev_test, criterion):
    stats = [(i+1, js[3*i+dev_test]) for i in range(len(js)//3)]
    sorted_by_stats = sorted(stats, key=lambda x:x[1][criterion], reverse=True)
    idx_by_criterion = [(it[0], it[1][criterion]) for it in sorted_by_stats]
    return idx_by_criterion
```
